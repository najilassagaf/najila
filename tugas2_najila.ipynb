```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = '/content/encoded_cleaned_test.csv'
df = pd.read_csv(dataset)
df
```

```python
# @title Delivery_person_Ratings

from matplotlib import pyplot as plt
df['Delivery_person_Ratings'].plot(kind='hist', bins=20, title='Delivery_person_Ratings')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

dari grafik diatas dapat dilihat bahwa delivery person ratings memiliki frekuensi paling tinggi di angka antara 4 dan 5 sebaliknya frekuensi paling rendah di angka antara 3 dan 4

```python
# @title Type_of_vehicle

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Type_of_vehicle').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

dari grafik diatas dapat dilihat type of vehicle yang paling banyak digunakan adalah tipe 2 sedangkan type of vehicle yang paling sedikit digunakan adalah bicycle
